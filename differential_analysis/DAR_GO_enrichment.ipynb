```r
library(rGREAT)
library(dplyr)
```

```r
peaks <- read.table("total_DAR_rgreat_input.bed", header = FALSE)
```

```r
run_great_go_by_group <- function(peaks, 
                                  cell_col = "V7",      # 细胞类型列
                                  contrast_col = "V5",  # 比较组列（如 "Underweight_vs_Normal"）
                                  dir_col = "V13",      # 方向列（"up" / "down"）
                                  pval_col = "V10",     # chr
                                  start_col = "V11",    # start
                                  end_col = "V12",      # end
                                  genome = "hg38",
                                  fdr_cutoff = 0.05) {
  
  # 提取唯一组合
  groups <- peaks %>%
    distinct(!!sym(cell_col), !!sym(contrast_col), !!sym(dir_col)) %>%
    filter(!is.na(!!sym(dir_col)))  # 排除 NA
  
  all_results <- list()
  
  for (i in seq_len(nrow(groups))) {
    ct   <- groups[[cell_col]][i]
    comp <- groups[[contrast_col]][i]
    dir  <- groups[[dir_col]][i]
    
    cat("Processing:", ct, "|", comp, "|", dir, "\n")
    
    # 筛选子集
    sub_peaks <- peaks %>%
      filter(!!sym(cell_col) == ct,
             !!sym(contrast_col) == comp,
             !!sym(dir_col) == dir) %>%
      select(chr = !!sym(pval_col),  # 注意：V11 是 chr 列？请确认！
             start = !!sym(start_col),
             end = !!sym(end_col))
    
    # 检查是否有 peak
    if (nrow(sub_peaks) == 0) {
      cat("  -> No peaks, skipping.\n")
      next
    }
    
    # 转 GRanges（确保列名为 seqnames/start/end）
    names(sub_peaks) <- c("seqnames", "start", "end")
    gr <- makeGRangesFromDataFrame(sub_peaks, ignore.strand = TRUE)
    
    # 提交 GREAT 任务
    set.seed(123)
    job <- tryCatch({
      submitGreatJob(gr, genome = genome)
    }, error = function(e) {
      cat("  -> GREAT submission failed:", e$message, "\n")
      return(NULL)
    })
    
    if (is.null(job)) next
    
    # 获取 GO Molecular Function 表
    tbl <- tryCatch({
      getEnrichmentTables(job)
    }, error = function(e) {
      cat("  -> Failed to retrieve results:", e$message, "\n")
      return(NULL)
    })
    
    if (is.null(tbl) || is.null(tbl$`GO Molecular Function`)) {
      cat("  -> No GO MF results.\n")
      next
    }
    
    go_df <- tbl$`GO Molecular Function`
    
    # 筛选显著结果
    sig_go <- go_df %>%
      filter(Binom_Adjp_BH < fdr_cutoff) %>%
      mutate(
        cell_type = ct,
        contrast = comp,
        direction = dir,
        .before = everything()
      )
    
    if (nrow(sig_go) > 0) {
      all_results[[length(all_results) + 1]] <- sig_go
    } else {
      cat("  -> No significant terms.\n")
    }
  }
  
  # 合并所有结果
  if (length(all_results) == 0) {
    stop("No significant GO terms found for any group.")
  }
  
  final_df <- bind_rows(all_results)
  return(final_df)
}
```

```r
result_all <- run_great_go_by_group(peaks)
```
